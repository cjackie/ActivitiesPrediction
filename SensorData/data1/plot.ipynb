```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
accelerometer_file = 'data/accelerometer_1.csv'
accelerometer = []
with open(accelerometer_file, 'r') as f:
    f.readline()
    for l in f:
        accelerometer.append(map(lambda s: float(s), l.split(',')))
accelerometer = np.array(accelerometer)
fig = plt.figure()
fig.add_subplot(1,1,1)
plt.title('accelerometer in x direction')
plt.plot(accelerometer[:,3], accelerometer[:,0])

fig = plt.figure()
fig.add_subplot(1,1,1)
plt.title('accelerometer in y direction')
plt.plot(accelerometer[:,3], accelerometer[:,1])

fig = plt.figure()
fig.add_subplot(1,1,1)
plt.title('accelerometer in z direction')
plt.plot(accelerometer[:,3], accelerometer[:,2])
```
